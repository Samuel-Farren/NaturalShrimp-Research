# Iowa Facility

```python
survival_rates = [0.9,0.95,0.99]
stocking_densities = [500,750,1000]
growth_rate = [0.8,1.1,1.3,1.6,2]
subsystem_capacity = 151
number_subsystems = [3,4,5,6,7]
```

```python
# Weekly Harvest weight for 18-22 Shrimp = Water Capacity used in Weekly harvest (cubic meters)* 
# Stocking Density(shrimp / cubic meters) * 
# Survival Rate / 20 shrimp per pound
```

```python
webster_city_df = pd.DataFrame(columns=["Survival Rate", "Stocking Density", "Growth Rate", "Number of Subsystems Used", "Gallons Used in Weekly Harvest", "Weekly Harvest Yield (Pounds)"])
harvest_weights = []
for survival in survival_rates:
    for density in stocking_densities:
        index = 0
        for growth in growth_rate:
            total = number_subsystems[index] * subsystem_capacity * density * survival * 20 / 453
            harvest_weights.append(total)
            print("Survival: ",survival, "  Density: ",density, "  Growth Rate: ", growth  ,"  Total: ",total, " pounds")
            webster_city_df = webster_city_df.append({
                 "Survival Rate": survival,
                 "Stocking Density":  density,
                 "Growth Rate": growth,
                 "Number of Subsystems Used": number_subsystems[index],
                 "Gallons Used in Weekly Harvest": number_subsystems[index] * 40000,
                 "Weekly Harvest Yield (Pounds)": total,
                }, ignore_index=True)
            index = index + 1
            
```

```python
webster_city_df.to_csv("WebsterCity_Projections.csv")
```
